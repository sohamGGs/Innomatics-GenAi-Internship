# Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = eval(input("List of integers : "))

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = eval(input("List of integers : "))

count = 0

for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = eval(input("Enter list of status(P / A) : "))

present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {attendance_percentage}")
```

# Problem 6: Remove Duplicate Phone Numbers

# Problem 7: Count Character Frequency

```python
text = input("Enter string : ")

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

# Problem 10: Calculate Average Marks

```python
marks = eval(input("Enter list of marks : "))

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print(f"Average Marks: {average}")
```
