# Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

def get_top_performers(data):
    if not data:
        return ""
    max_score = max(data.values())
    top_list = [name for name, score in data.items() if score == max_score]
    return f"Top Performers Eligible for Bonus: {', '.join(top_list)} (Score: {max_score})"

print(get_top_performers(employees))
```

# Problem Statement 2: Search Query Keyword Analysis

```python
def analyze_keywords(query):
    query = query.lower()

    punctuation = ".,!?;:"
    for char in punctuation:
        query = query.replace(char, "")

    words = query.split()
    word_counts = {}

    for word in words:
        word_counts[word] = word_counts.get(word, 0) + 1

    frequent_keywords = {word: count for word, count in word_counts.items() if count > 1}

    return frequent_keywords

input_query = "Buy mobile phone buy phone online"
print(analyze_keywords(input_query))
```

# Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

def validate_sensor_data(readings):
    valid_readings = [(index, value) for index, value in enumerate(readings) if value % 2 == 0]

    print("Valid Sensor Readings (Hour, Value):")
    print(valid_readings)

validate_sensor_data(sensor_readings)
```

# Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

def analyze_email_domains(email_list):
    domain_counts = {}
    total_emails = len(email_list)

    for email in email_list:
        domain = email.split('@')[-1]
        domain_counts[domain] = domain_counts.get(domain, 0) + 1

    for domain, count in domain_counts.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {int(percentage)}%")

analyze_email_domains(emails)
```

# Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

def detect_sales_spikes(sales_list):
    if not sales_list:
        return

    avg_sales = sum(sales_list) / len(sales_list)
    threshold = avg_sales * 1.3

    for index, value in enumerate(sales_list, start=1):
        if value > threshold:
            print(f"Day {index}: {value}")

detect_sales_spikes(sales)
```

# Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

def detect_duplicates(ids):
    counts = {}
    for uid in ids:
        counts[uid] = counts.get(uid, 0) + 1

    for uid, count in counts.items():
        if count > 1:
            print(f"{uid} → {count} times")

detect_duplicates(user_ids)
```
