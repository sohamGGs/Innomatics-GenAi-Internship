# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_data):

    VIRAL_THRESHOLD = 1000
    total_likes = 0

    # Step 1: Use a loop to calculate total likes (as per hint)
    for likes in likes_data:
        total_likes += likes

    # Step 2: Apply conditional logic to determine status
    if total_likes >= VIRAL_THRESHOLD:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: Likes received from different sources/time slots
    post_metrics = [250, 400, 150, 250]

    # Unpacking the returned tuple
    total, post_status = analyze_post_engagement(post_metrics)

    # Expected Output Format
    print(f"Total Likes: {total}")
    print(f"Post Status: {post_status}")
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall_adequacy(rainfall_readings, required_level):

    if not rainfall_readings:
        return 0, "No data available"

    total_rainfall = 0

    # Step 1: Use a loop to calculate total rainfall (as per hint)
    for amount in rainfall_readings:
        total_rainfall += amount

    # Step 2: Calculate average using math operations
    average_rainfall = total_rainfall / len(rainfall_readings)

    # Step 3: Determine adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return int(average_rainfall), status

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: Rainfall in mm over 4 months
    seasonal_data = [80, 65, 90, 55]
    requirement = 70

    avg, rainfall_status = check_rainfall_adequacy(seasonal_data, requirement)

    # Expected Output Format
    print(f"Average Rainfall: {avg}")
    print(f"Rainfall Status: {rainfall_status}")
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicates(username_list):

    # Step 1: Use a set to find unique usernames
    # A set automatically removes duplicates from the original list
    unique_usernames = set(username_list)

    # Step 2: Compare lengths to detect if any duplicates were removed
    if len(username_list) != len(unique_usernames):
        return "Yes"
    else:
        return "No"

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: A list of usernames trying to register
    database_usernames = ["user_101", "dev_soham", "alpha_user", "user_101", "tech_lead"]

    # Call the function
    result = detect_duplicates(database_usernames)

    # Expected Output Format
    print(f"Duplicate Accounts Found: {result}")
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(price_list):

    PREMIUM_THRESHOLD = 2000
    premium_crops = []

    # Step 1: Use list filtering logic (via a loop) to identify premium prices
    for price in price_list:
        if price > PREMIUM_THRESHOLD:
            # Step 2: Add premium prices to the new list
            premium_crops.append(price)

    return premium_crops

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: Prices of different crops in the market
    market_prices = [1500, 2500, 1800, 3200, 900, 4500]

    # Call the function
    premium_list = filter_premium_crops(market_prices)

    # Expected Output Format
    print(f"Premium Crops: {premium_list}")
```

# Problem 7: System Monitoring – Application Health Checker

```python
def check_system_health(error_count):

    # Rule 1: Zero errors means the system is fully operational
    if error_count == 0:
        status = "Healthy"

    # Rule 2: Moderate errors indicate a need for investigation
    elif error_count <= 5:
        status = "Minor Issues"

    # Rule 3: High error counts require immediate intervention
    else:
        status = "Critical Issues"

    return status

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: A system reporting 7 errors in the last minute
    current_errors = 7

    # Call the function
    health_status = check_system_health(current_errors)

    # Expected Output Format
    print(f"Error Count: {current_errors}")
    print(f"System Status: {health_status}")
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_log):

    if not attendance_log:
        return 0.0, "No data"

    total_classes = len(attendance_log)
    days_present = 0

    # Step 1: Use a loop to count present days (as per hint)
    for status in attendance_log:
        if status == 1:
            days_present += 1

    # Step 2: Calculate percentage
    percentage = (days_present / total_classes) * 100

    # Step 3: Check eligibility against the 75% threshold
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return round(float(percentage), 1), status

# --- Main Execution ---
if __name__ == "__main__":
    # Sample data: 1 = Present, 0 = Absent
    # Student attended 4 out of 5 classes
    student_log = [1, 1, 0, 1, 1]

    percent, eligibility = check_exam_eligibility(student_log)

    # Expected Output Format
    print(f"Attendance Percentage: {percent}")
    print(f"Exam Eligibility: {eligibility}")
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):

    total_bill = 0

    # Step 1: Calculate Bill based on Slabs (Bucket Logic)
    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        # First 100 units at ₹3 + remaining units at ₹5
        total_bill = (100 * 3) + ((units - 100) * 5)
    else:
        # First 100 at ₹3 + next 100 at ₹5 + remainder at ₹7
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Step 2: Classify Usage Level
    if total_bill < 500:
        usage_status = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return float(total_bill), usage_status

# --- Main Execution ---
if __name__ == "__main__":
    # Sample: A household consumes 250 units
    units_consumed = 250

    bill_amount, status = calculate_electricity_bill(units_consumed)

    # Expected Output Format
    print(f"Units Consumed: {units_consumed}")
    print(f"Total Bill: ₹{bill_amount}")
    print(f"Usage Status: {status}")
```
