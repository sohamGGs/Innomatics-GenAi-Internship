# 1. User Login Check

```python
username = "admin"
password = "1234"

user_match = False

name = input("Username: ")
code = input("Password: ")

if name == username:
    if code == password:
        user_match = True

if user_match:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_c = 0
fail_c = 0

for m in marks:
    if m >= 50:
        pass_c += 1
    else:
        fail_c += 1

print("Total Pass:", pass_c)
print("Total Fail:", fail_c)
```

# 3. Simple Data Cleaner

```python
names = ["  Alice  ", "bob", "  CHARLIE  "]
clean_list = []

for n in names:
    clean_list.append(n.strip().lower())

print(clean_list)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    size = len(msg)
    print("Length:", size)

    if size > 10:
        print("Flagged: Message too long")
```

# 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total error count:", error_count)
```
