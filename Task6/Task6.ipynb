# Problem Statement 1:Smart Parking Lot Management System

```python
def manage_parking_lot():
    # Dynamic Input for Capacity
    try:
        capacity = int(input("Enter Parking Capacity: "))
    except ValueError:
        print("Invalid input. Please enter a number for capacity.")
        return

    # Dynamic Input for Vehicle Logs
    # Enter logs separated by spaces (e.g., IN IN OUT IN)
    logs_input = input("Enter Vehicle Logs (IN/OUT separated by space): ").upper()
    vehicle_logs = logs_input.split()

    current_parked = 0
    peak_usage = 0

    # Processing logs
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            # Ensure parked count doesn't go below zero
            if current_parked > 0:
                current_parked -= 1

        # Track the peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Output Results
    print(f"\nCurrently Parked Vehicles: {current_parked}")

    # Logic for Alerts and Status
    if current_parked >= capacity:
        print("Parking Status: FULL (Alert: Capacity Exceeded)")
    else:
        print("Parking Status: Available")

    print(f"Peak Usage during this period: {peak_usage}")

# Run the function
if __name__ == "__main__":
    manage_parking_lot()
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time():
    print("--- Online Food Delivery Time Estimator ---")

    # Dynamic Inputs
    try:
        distance = float(input("Distance (km): "))
        traffic = input("Traffic Level (Low/Medium/High): ").strip().capitalize()
        weather = input("Weather (Clear/Rainy/Snowy): ").strip().capitalize()
    except ValueError:
        print("Invalid input. Please enter a numerical value for distance.")
        return

    # 1. Base Time calculation (e.g., 4 minutes per km)
    base_time = distance * 4

    # 2. Traffic Delays
    traffic_delay = 0
    if traffic == "High":
        traffic_delay = 15
    elif traffic == "Medium":
        traffic_delay = 8
    else:
        traffic_delay = 0

    # 3. Weather Delays
    weather_delay = 0
    if weather == "Rainy":
        weather_delay = 8
    elif weather == "Snowy":
        weather_delay = 15
    else:
        weather_delay = 0

    # Final ETA Calculation
    # We add a constant 5-minute buffer for food preparation
    total_eta = int(base_time + traffic_delay + weather_delay)

    print("-" * 30)
    print(f"Estimated Delivery Time: {total_eta} minutes")

if __name__ == "__main__":
    estimate_delivery_time()
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_movie_theatre():
    print("--- Movie Theatre Seat Occupancy Analyzer ---")

    try:
        # Dynamic Input for Total Seats
        total_seats = int(input("Total Seats: "))

        # Dynamic Input for Booked Seats
        # In this logic, we assume each '1' in the list represents a booking
        # User can enter 150 ones separated by spaces, or any sequence
        booked_input = input("Enter Booked Seats (e.g., 1 1 1...): ").split()
        booked_count = len(booked_input)

    except ValueError:
        print("Invalid input. Please enter numbers only.")
        return

    # 1. Occupancy Percentage Calculation
    # Formula: (Booked / Total) * 100
    occupancy_pct = (booked_count / total_seats) * 100

    # 2. Determine Status and Suggestions
    if occupancy_pct == 100:
        status = "Housefull"
        suggestion = "Opening additional shows is highly recommended."
    elif occupancy_pct >= 70:
        status = "Almost Full"
        suggestion = "Consider opening an extra show for high demand."
    elif occupancy_pct >= 50:
        status = "Moderately Full"
        suggestion = "Monitor for further bookings."
    else:
        status = "Available"
        suggestion = "No additional shows needed."

    # Output Results
    print("-" * 30)
    print(f"Occupancy: {int(occupancy_pct)}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")

if __name__ == "__main__":
    analyze_movie_theatre()
```

# Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load():
    print("--- Cloud Server Load Classification System ---")

    # Dynamic Input for CPU Readings
    try:
        # User enters numbers separated by spaces
        raw_input = input("Enter CPU Readings (%) separated by spaces: ")
        cpu_readings = [float(x) for x in raw_input.split()]

        if not cpu_readings:
            print("No readings provided.")
            return

        # 1. Average Calculation
        # Formula: (Sum of all readings) / (Number of readings)
        avg_cpu = sum(cpu_readings) / len(cpu_readings)

        # 2. Status Classification
        if avg_cpu >= 80:
            status = "Critical"
        elif 50 <= avg_cpu < 80:
            status = "Warning"
        else:
            status = "Normal"

        # Output Results
        print("-" * 30)
        print(f"Average CPU Load: {int(avg_cpu)}%")
        print(f"Server Status: {status}")

    except ValueError:
        print("Invalid input. Please enter numbers only.")

if __name__ == "__main__":
    classify_server_load()
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_classroom_resources():
    print("--- Smart Classroom Resource Usage Monitor ---")

    # Dynamic Input for Resources
    resources = {}
    print("Enter the usage hours for the following resources:")

    try:
        resources["Projector"] = float(input("Projector usage (hours): "))
        resources["AC"] = float(input("AC usage (hours): "))
        resources["Lights"] = float(input("Lights usage (hours): "))
    except ValueError:
        print("Invalid input. Please enter numbers only.")
        return

    # 1. Identify Overuse (Threshold = 8 hours)
    overused = []
    threshold = 8

    for item, hours in resources.items():
        if hours > threshold:
            overused.append(item)

    # 2. Output Logic
    print("-" * 30)
    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

if __name__ == "__main__":
    monitor_classroom_resources()
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registrations():
    print("--- Online Event Registration Capacity Controller ---")

    try:
        # Dynamic Input
        capacity = int(input("Event Capacity: "))
        total_applicants = int(input("Registrations: "))

        # 1. Logic for Confirmation and Waitlist
        if total_applicants <= capacity:
            confirmed = total_applicants
            waitlist = 0
            status = "Open"
        else:
            confirmed = capacity
            waitlist = total_applicants - capacity
            status = "Closed"

        # 2. Output Results
        print("-" * 30)
        print(f"Confirmed Registrations: {confirmed}")
        print(f"Waitlisted Users: {waitlist}")
        print(f"Registration Status: {status}")

    except ValueError:
        print("Invalid input. Please enter whole numbers for capacity and registrations.")

if __name__ == "__main__":
    manage_event_registrations()
```
